```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Toy dataset from scikit-learn

```python
from sklearn.datasets import make_blobs

X, *_ = make_blobs(n_samples=10_000, n_features=2)
```

We can use farthest point sampling both at the function level and at the class level

```python
from moldex.sample_selection import farthest_point_sampling

centers, landmarks = farthest_point_sampling(X, n_centers=50)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4), dpi=100)
ax.scatter(X[:, 0], X[:, 1], s=1, label='Data')
ax.scatter(X[centers, 0], X[centers, 1], s=20, ec='k', label='Sampled landmarks')
ax.legend()
```
